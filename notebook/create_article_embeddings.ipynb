# load article data

```python
import numpy as np
import pandas as pd
import os
pd.set_option('display.max_columns', None)
```

```python
data_dir = "/content"
save_dir = "/content/drive/MyDrive/recsys24_share"
article = pd.read_parquet(os.path.join(data_dir, '/articles.parquet'))
article["title_subtitle"] = article["title"] + "\n" + article["subtitle"]
```

# create title embeddings

## model preparation

```python
from sentence_transformers import SentenceTransformer
model_name = "distiluse-base-multilingual-cased-v2"
model = SentenceTransformer(model_name)
```

## create embeddings

```python
def extract_embeddings(column_name, df=article, model_name=model_name, model=model):
    sentences = article[column_name].to_list()
    sentence_embeddings = model.encode(sentences)
    np.save(os.path.join(save_dir, f"article_{column_name}_embeddings_{model_name}"), sentence_embeddings)
```

```python
extract_embeddings("title")
extract_embeddings("subtitle")
extract_embeddings("title_subtitle")
```
